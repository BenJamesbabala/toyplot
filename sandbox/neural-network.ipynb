Minimum requirements to draw neural network diagrams:

* Layered layout algorithm
* Render edge arrows
    * Unfortunately, they're obscured by markers - we'd have to figure-out where the edge intersects the marker to do it up right.
* Draw boxes around layers

```python
import numpy
import toyplot.color
import toyplot.generate
import toyplot.mark
```

```python
edges = numpy.array([
    ["x0", "a0"],
    ["x0", "a1"],
    ["x0", "a2"],
    ["x0", "a3"],
    ["x1", "a0"],
    ["x1", "a1"],
    ["x1", "a2"],
    ["x1", "a3"],
    ["x2", "a0"],
    ["x2", "a1"],
    ["x2", "a2"],
    ["x2", "a3"],
])

vcoordinates = numpy.ma.masked_all((7, 2))
vcoordinates[0] = (0, 0)
vcoordinates[1] = (1, 0)
vcoordinates[2] = (2, 0)
vcoordinates[3] = (3, 0)
vcoordinates[4] = (0, 1)
vcoordinates[5] = (1, 1)
vcoordinates[6] = (2, 1)
vcoordinates[:, 0] *= 0.4

canvas = toyplot.Canvas(width=1000, height=500)

axes = canvas.cartesian(show=False, padding=30)
axes.aspect = "fit-range"
mark = axes.graph(
    edges,
    vcoordinates=vcoordinates,
    vcolor="white",
    vstyle={"stroke":"black"},
    vlshow=False,
    vsize=20,
    vmarker=None,
    ecolor="black",
    eopacity=1,
    estyle={"stroke-width":1},
    tmarker=toyplot.marker.create(">", mstyle={"fill":"white", "stroke-width":1}, size=15),
)
#mark = axes.rects(-0.125, 0.8 + 0.125, 1 - 0.125, 1 + 0.125, color="rgba(0, 0, 0, 0.2)")
```
