```python
import csv
import os
import numpy
import toyplot.data
```

```python
rows = [row for row in csv.reader(open("co2-pop.csv"))]
```

```python
columns = {unicode(row[0].strip(), encoding="utf8") : numpy.array(row[1:]) for row in rows if row[0].strip() not in ["", "CO2 emissions / population", "Back to the table of contents"]}
```

```python
for name, column in columns.items():
    column[column == ""] = "0"
    column[column == ".."] = "0"
    column[column == "x"] = "0"
    column = column.astype("float64")
    column = numpy.ma.masked_where(column == 0, column)
    columns[name] = column
```

```python
year = columns["tonnes CO2 / capita"]
```

```python
del columns["tonnes CO2 / capita"]
del columns["Region/Country/Economy"]
```

```python
toyplot.data.Table(columns)
```

```python
canvas = toyplot.Canvas()
axes = canvas.axes()
for key in sorted(columns.keys()):
    axes.plot(year, columns[key])
```

```python
canvas = toyplot.Canvas()
axes = canvas.axes(ymin=0)
for key in ["Former Soviet Union (if no detail)", "Russian Federation", "Former Soviet Union"]:
    axes.plot(year, columns[key])
```

```python
canvas = toyplot.Canvas()
axes = canvas.axes()
for key in sorted(columns.keys()):
    series = columns[key]
    first = series.compressed()[0]
    axes.plot(year, series - first)
```

```python
deltas = numpy.array([columns[key].compressed()[-1] - columns[key].compressed()[0] for key in sorted(columns.keys())])
sort_order = numpy.argsort(deltas)
```

```python
toyplot.bars(deltas[sort_order])
```

```python
data = toyplot.data.Table()
data["Entity"] = numpy.array(sorted(columns.keys()))
data["CO2 tons / capita"] = [""] * len(data)
data
```

```python
canvas = toyplot.Canvas(width=400, height=4000)
table = canvas.table(data)
table.column(1).width = 100
for index, key in enumerate(sorted(columns.keys())):
    column = columns[key]
    deltas = column[1:] - column[:-1]
    #deltas[deltas < 0] = -1
    #deltas[deltas > 0] = 1
    fill = numpy.tile("green", len(deltas))
    fill[deltas > 0] = "red"
    #axes.cell(index, 1).axes().bars(deltas, fill=fill)
    stroke = "red" if column.compressed()[-1] > column.compressed()[0] else "black"
    table.cell(index, 1).axes().plot(column, stroke=stroke, style={"stroke-width":0.5})
```
