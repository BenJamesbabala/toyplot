# Data

```python
import toyplot.data
import numpy
```

```python
table = toyplot.data.read_csv("temperatures.csv")
table[0:10]
```

```python
max_f = ((table["TMAX"].astype("float64") * 0.1) * 1.8) + 32
min_f = ((table["TMIN"].astype("float64") * 0.1) * 1.8) + 32

canvas = toyplot.Canvas(width=600, height=300)
axes = canvas.axes(xlabel="Day", ylabel=u"Temperature \u00b0F")
axes.plot(max_f, color="red", stroke_width=1)
axes.plot(min_f, color="blue", stroke_width=1)
```
