```python
import numpy
import toyplot
```

```python
x = numpy.linspace(0, 2 * numpy.pi, 51)
m1 = numpy.ma.column_stack((
    1 + 0.5 * numpy.sin(x),
    1 + 0.5 * numpy.cos(x),
    1 + 0.2 * numpy.sin(2 * x),
    ))
m2 = numpy.ma.copy(m1)

m2[8:18, 0] = numpy.nan
m2[33:43, 1] = numpy.ma.masked
```

# Plots / Scatterplots

```python
toyplot.plot(m1, marker="o", size=40, width=400, height=400)
toyplot.plot(m2, marker="o", size=40, width=400, height=400)
```

# Magnitudes

```python
toyplot.bars(m1, width=400, height=400)
toyplot.bars(m2, width=400, height=400)
```

```python
toyplot.fill(m1, baseline="stacked", width=400, height=400)
toyplot.fill(m2, baseline="stacked", width=400, height=400)
```

# Boundaries

```python
numpy.random.seed(1234)
observations = numpy.random.normal(size=(50, 50))

b1 = numpy.ma.column_stack((numpy.min(observations, axis=1), numpy.median(observations, axis=1), numpy.max(observations, axis=1)))
b2 = numpy.ma.copy(b1)
b2[10:20, 0] = numpy.nan
b2[30:40, 1] = numpy.ma.masked
b2[20:30, 2] = numpy.nan
```

```python
toyplot.bars(b1, baseline=None, width=400, height=400)
toyplot.bars(b2, baseline=None, width=400, height=400)
```

```python
toyplot.fill(b1, width=400, height=400)
toyplot.fill(b2, width=400, height=400)
```
