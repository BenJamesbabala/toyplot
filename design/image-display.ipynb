```python
import skimage.color
import skimage.data
import skimage.filters
import toyplot
import PIL
```

```python
image = skimage.data.astronaut()
toyplot.image(image, width=300);
```

```python
gray = skimage.color.rgb2gray(image)
toyplot.image(gray, width=300);
```

```python
binary = gray > skimage.filters.threshold_isodata(gray)
toyplot.image(binary, width=300);
```
