```python
import numpy
import toyplot
```

```python
data="""15feb2005 553 1.0000
15mar2005 507 1.0000
15apr2005 586 1.0000
15may2005 488 1.0000
15jun2005 404 1.0000
15jul2005 306 1.0000
15aug2005 323 0.9905
15sep2005 531 0.9959
15oct2005 677 0.9600
15nov2005 695 0.9624
15dec2005 867 0.9229
15jan2006 557 0.9888"""
```

```python
rows = [row.split() for row in data.split("\n")]
date, delivered, on_time = zip(*rows)
```

```python
x = numpy.arange(len(date))
delivered = numpy.array(delivered).astype("float64")
on_time = numpy.array(on_time).astype("float64")
```

```python
canvas = toyplot.Canvas(width=1000, height=500)
axes = canvas.axes(xlabel="Date", ylabel="Delivered", ymin=0)
axes.plot(x, delivered, color="darkred", marker="o")
axes.y.spine.style = {"stroke":"darkred"}

axes = axes.share("x", label="On-time")
axes.plot(x, on_time, color="steelblue", marker="o")
axes.y.spine.style = {"stroke":"steelblue"}
axes.y.domain.min = 0
```
