```python
import numpy
import toyplot
```

```python
x = numpy.linspace(0, 10, 200)
y1 = numpy.sin(x)
y2 = x ** 2
```

```python
canvas = toyplot.Canvas()
axes = canvas.axes(xlabel="x", ylabel="sin(x)", gutter=100)
axes.x.ticks.show = True
axes.y.ticks.show = True
axes.plot(x, y1)

axes = axes.share("x", label="x^2")
axes.y.ticks.show = True
axes.plot(x, y2)
```
