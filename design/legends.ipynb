```python
import numpy
import toyplot

toyplot.log.setLevel(toyplot.logging.DEBUG)
```

```python
series = numpy.column_stack((numpy.linspace(0, 1), numpy.linspace(0, 1)**2))

canvas, axes, mark = toyplot.plot(series, width=400)

legend = [("A", mark)]
legend.append(("B", axes.plot(numpy.linspace(0, 1)**3, style={"stroke-dasharray":"2,2"})))

canvas.legend(legend, corner=("top-left", 50, 100, 70), style={"fill":"#f4f4f4"});
```
