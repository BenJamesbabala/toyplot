```python
import numpy
import toyplot

toyplot.log.setLevel(toyplot.logging.DEBUG)
```

```python
x = numpy.linspace(0, 2 * numpy.pi, 200)
canvas = toyplot.Canvas(800, 600)
axes = canvas.cartesian()
plots = [axes.plot(x, 0.5 * i * numpy.sin(x * i))
         for i in numpy.linspace(1, 2, 3)]
fill = axes.fill(x, numpy.sin(x) * 0.1, numpy.cos(x) * 0.1)
axes.x.label.text = "Time"
axes.y.label.text = "Temp"

legend = [
    ("Plot 1", plots[0]),
    ("Plot 2", plots[1]),
    ("Plot 3", plots[2]),
    ("Fill", fill),
    ("Explicit Line", {"shape":"/", "mstyle":{"stroke": "red", "stroke-width": 1.0}}),
    ("Explicit Rect", {"shape": "s", "mstyle":{"fill": "lightgreen"}}),
    ("Explicit Marker", {"shape":"^", "mstyle":{"fill": "lightblue", "stroke": toyplot.color.near_black}}),
    ("Explicit Marker", {"shape":"o", "mstyle":{"fill": "yellow", "stroke": toyplot.color.near_black}}),
    ("Explicit Marker", {"shape":"s", "mstyle":{"fill": "pink", "stroke": toyplot.color.near_black}}),
    ]

canvas.legend(legend, bounds=(100, 200, 350, 550))
```

```python
x =["2014-01", "2014-02", "2014-03", "2014-04", "2014-05", "2014-06", "2014-07", "2014-08", "2014-09", "2014-10", "2014-11", "2014-12"]
a = [0.43, 0.37, 0.33, 0.3, 0.28, 0.25, 0.22, numpy.nan, numpy.nan, numpy.nan, numpy.nan, numpy.nan]
b = [numpy.nan, numpy.nan, numpy.nan, numpy.nan, numpy.nan, numpy.nan, numpy.nan, 0.21, 0.19, 0.10, 0.07, 0.03]
```

```python
canvas = toyplot.Canvas(width=600, height=400)
axes = canvas.cartesian(label='Hard Drive Failures', xlabel="Vehicle Build Month", ylabel="Defect Rate (%)", gutter=100)
legend = []
legend.append(("A", axes.bars(a)))
legend.append(("B", axes.bars(b)))
legend.append(("A", axes.plot(a, color="black")))
legend.append(("B", axes.scatterplot(b, marker="oo", size=15, mstyle={"stroke":"black"})))

canvas.legend(legend, corner=("top-right", 90, 50, 22*4))
    
axes.x.ticks.locator = toyplot.locator.Explicit(labels=x)
axes.x.ticks.labels.angle=90
axes.x.label.style = {"baseline-shift":"-70px"}
```
