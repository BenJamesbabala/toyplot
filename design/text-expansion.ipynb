```python
import numpy
import toyplot
```

```python
x = numpy.linspace(0, 1, 10)
y = x ** 2
text = numpy.repeat("Text", len(x))
```

```python
canvas = toyplot.Canvas()
axes = canvas.axes()
#axes.scatterplot(x, y, color="black")
axes.text(x, y, text, annotation=False)
```

```python
canvas = toyplot.Canvas()
axes = canvas.axes()
axes.text(0, 0, "Text", annotation=True)
```
