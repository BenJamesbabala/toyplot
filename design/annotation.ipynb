```python
import numpy
x = numpy.arange(10)
y1 = numpy.random.normal(size=len(x))
y2 = numpy.random.normal(size=len(x))
y3 = numpy.random.normal(size=len(x))
```

```python
import toyplot
canvas = toyplot.Canvas()
axes = canvas.axes()
axes.plot(x, y1)
axes.text(x[-1], y1[-1], "Label 1", angle=0, style={"text-anchor":"start", "-toyplot-anchor-shift":"5px"})
axes.plot(x, y2)
axes.text(x[-1], y2[-1], "Label 2", angle=0, style={"text-anchor":"start", "-toyplot-anchor-shift":"5px"})
axes.plot(x, y3)
axes.text(x[-1], y3[-1], "Label 3", angle=0, style={"text-anchor":"start", "-toyplot-anchor-shift":"5px"})
```
