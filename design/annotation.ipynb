```python
import numpy
numpy.random.seed(1234)
x = numpy.arange(100)
y1 = numpy.cumsum(numpy.random.normal(loc=0.05, size=len(x)))
y2 = numpy.cumsum(numpy.random.normal(loc=0.05, size=len(x)))
y3 = numpy.cumsum(numpy.random.normal(loc=0.05, size=len(x)))
```

```python
label_style={"text-anchor":"start", "-toyplot-anchor-shift":"5px"}

import toyplot
canvas = toyplot.Canvas()
axes = canvas.axes()
axes.plot(x, y1)
axes.text(x[-1], y1[-1], "Label Numero Uno", angle=0, style=label_style)
axes.plot(x, y2)
axes.text(x[-1], y2[-1], "The really long second label.", angle=-30, style=label_style)
axes.plot(x, y3)
axes.text(x[-1], y3[-1], "Label 3", angle=0, style=label_style)
```
