```python
import numpy
import toyplot
```

```python
numpy.random.seed(1234)
x = numpy.arange(100)
y = numpy.column_stack([numpy.cumsum(numpy.random.normal(loc=0.05, size=len(x))) for i in range(5)])
```

```python
canvas = toyplot.Canvas()
axes = canvas.axes()
mark = axes.plot(x, y, title=[1, 2, 3, 4, 5])
for i in range(5):
    axes.text(x[-1], y[-1,i], "Series %s" % i, style={"text-anchor":"start", "-toyplot-anchor-shift":"5px"})
```
