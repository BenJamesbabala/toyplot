```python
import numpy
import toyplot
```

```python
x = numpy.array([   32,    64,   128,   256,   512,  1024,  2048,  4096,  8192, 16384])
y = numpy.array([[numpy.nan,  64.28238627],
       [ numpy.nan,  35.11086749],
       [ 12.66786078,  19.32944274],
       [  7.74291979,  10.07842095],
       [  4.00576462,   5.11424115],
       [  1.7207776 ,   3.05380224],
       [  1.11896422,   2.20509594],
       [  0.57893869,   1.90313702],
       [  0.27078975,   1.25208092],
       [  0.16423923,    numpy.nan]])

canvas = toyplot.Canvas(400, 300)
axes = canvas.axes(xscale='log2', yscale='log10')
axes.x.ticks.locator = toyplot.locator.Explicit(2 ** numpy.arange(5,14))
axes.y.ticks.locator = toyplot.locator.Explicit(10 ** numpy.arange(-1,2))
axes.plot(x, y, marker='o', size=40)
```
