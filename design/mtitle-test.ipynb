```python
import numpy
import toyplot
```

```python
canvas, axes, mark = toyplot.plot(numpy.linspace(0, 1) ** 2, marker="o", title="a", mtitle="b")
```

```python
canvas = toyplot.Canvas()
table = canvas.table(numpy.random.normal(size=(5, 5)))
table.cells.row[2].cartesian().cell_plot(series="rows", marker="o", title="a", mtitle="b")
```
