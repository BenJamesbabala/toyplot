```python
import numpy
import toyplot.pdf
```

```python
x = numpy.zeros(10)
y = x
angle = numpy.linspace(90, 0, len(x), endpoint=True)
fill = numpy.linspace(1, 0, len(x))
```

```python
canvas = toyplot.Canvas(400, 400)
axes = canvas.axes(xmin=-0.25, xmax=0.5, ymin=-0.5, ymax=0.25)
mark = axes.text(x, y, text="Toyplot!", angle=angle, fill=fill, style={"font-size":"36px", "font-weight":"bold", "stroke":"white", "text-anchor":"begin"})
```

```python
toyplot.pdf.render(canvas, "test.pdf")
```
