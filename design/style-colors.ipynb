```python
import numpy as np
import toyplot

import logging
logging.basicConfig()

## generate histogram data
dat = np.random.normal(0, 1, 1000)
m, x = np.histogram(dat, bins=20)

## set up plot
c = toyplot.Canvas(width=300, height=300)
a = c.axes()

## raises an error
#color = toyplot.color.Palette()[1]
#heights = np.column_stack([[140 for i in m], m])
#a.bars(x[:-1], heights, 
#       color=['white', color],
#       baseline="stacked",
#       style={"stroke": color},  ## <--- doesn't like color as matrix
#      )

## this works
color = toyplot.color.Palette()[1]
heights = np.column_stack([[140 for i in m], m])
a.bars(x[:-1], heights, 
       color=['white', color],
       baseline="stacked",
       style={"stroke": toyplot.color.to_css(color)},  
      )
```
