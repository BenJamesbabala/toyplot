```python
import numpy
import toyplot
```

```python
matrix = numpy.random.normal(size=(20, 20))
print matrix.min(), matrix.max()
```

```python
colormap = toyplot.color.LinearMap(toyplot.color.brewer("BlueRed"), -3, 3)
canvas = toyplot.Canvas()
canvas.matrix(matrix, title="A matrix", colormap=colormap)
```
