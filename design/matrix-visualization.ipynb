```python
import numpy
import toyplot
```

Create some sample data using a normal distribution:

```python
numpy.random.seed(1234)
matrix = numpy.random.normal(size=(20, 20))
```

By default, matrix() uses a diverging linear colormap based on the data min/max:

```python
canvas = toyplot.Canvas()
canvas.matrix(matrix, title="A matrix")
```

But, you can use your own colormap if you prefer - in this case, a different palette, and adjusted so that it's symmetric around zero:

```python
print matrix.min(), matrix.max()
```

```python
colormap = toyplot.color.LinearMap(toyplot.color.brewer("PurpleOrange"), domain_min = -3, domain_max = 3)
```

```python
canvas = toyplot.Canvas()
canvas.matrix(matrix, title="A matrix", colormap=colormap)
```
