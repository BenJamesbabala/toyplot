```python
import numpy
import toyplot
```

```python
canvas, axes, mark = toyplot.plot(numpy.linspace(0, 1) ** 2, width=600, height=600 / 1.6, canvas_style={"background-color":"white"})
```

```python
import toyplot.png
toyplot.png.render(canvas, "normal.png")
toyplot.png.render(canvas, "width.png", width=450)
toyplot.png.render(canvas, "height.png", height=100)
toyplot.png.render(canvas, "scale.png", scale=2.0)
```

```python
import toyplot.pdf
toyplot.pdf.render(canvas, "normal.pdf")
toyplot.pdf.render(canvas, "width.pdf", width=(4.5, "inches"))
toyplot.pdf.render(canvas, "height.pdf", height=(20, "cm"))
toyplot.pdf.render(canvas, "scale.pdf", scale=2.0)
```

```python
import toyplot.eps
toyplot.eps.render(canvas, "normal.eps")
toyplot.eps.render(canvas, "width.eps", width=(4.5, "inches"))
toyplot.eps.render(canvas, "height.eps", height=(20, "cm"))
toyplot.eps.render(canvas, "scale.eps", scale=2.0)
```
