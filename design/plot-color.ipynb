# Line/Scatterplot plot color behavior

* default-color: per-series

*Lines*

* color: per-series, overrides default-color
* stroke-width: per-series
* stroke-opacity: per-series
* style: singular, overrides color, stroke-width, stroke-opacity

*Markers*

* msize: per-datum
* mfill: per-datum, overrides color
* mstroke: per-datum, overrides color
* mopacity: per-datum
* mstyle: singular, overrides mfill, mstroke

*Marker Labels*

* mlstyle: singular

```python
import numpy
import toyplot
```

```python
x = numpy.linspace(0, 1)
```

```python
canvas = toyplot.Canvas()
axes = canvas.axes()
mark = axes.plot(x, numpy.column_stack((x**2, x**3, x**4)), marker=["x", "o", "s"], size=64)
```

```python
canvas = toyplot.Canvas()
axes = canvas.axes()
mark = axes.plot(x, numpy.column_stack((x**2, x**3, x**4)), color=["red", "green", "blue"], marker=["x", "o", "s"], fill="white", size=64)
```

```python
canvas = toyplot.Canvas()
axes = canvas.axes()
mark = axes.scatterplot(x, numpy.column_stack((x**2, x**3, x**4)), marker=["x", "o", "s"], size=64)
```
