```python
import numpy
import toyplot
```

```python
canvas = toyplot.Canvas()
axes = canvas.axes()
for i in range(8):
    x = numpy.random.normal(loc=i, size=1000)
    y = numpy.random.normal(loc=i, size=1000)
    axes.scatterplot(x, y, size=40, mstyle={"opacity":0.4, "stroke":"none", "stroke-width":0.1})
```
