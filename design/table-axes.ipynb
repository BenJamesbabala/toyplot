### Separate Region Pros

* Changing header dimensions doesn't affect body cell indexing.
* Changing header dimensions doesn't affect body grid indexing.
* Body cell indexing matches data.

### Separate Region Cons

* Header cells must be offset / styled / aligned separately from body cells.
    * Workaround: add a *header* argument to the *column* API that causes it to alter both header and body cells.
        * *show*, *format*, *align*, *style*, *width*, *height*, *column_offset*, and *row_offset* would work in the obvious way.
        * What do we do for *data* in this case?  Does it only apply to body cells?
        * *merge()* wouldn't be available in this case, since we can't merge across regions.
        * *axes()* could still be allowed.
* Header grid lines must be specified separately from body grid lines.
    * Workaround: ???
* Overlapping grid lines are confusing.
    * Workaround: bordering grid lines are forced to stay in-sync.

### Consolidated Region Pros

* Header and body cells can be offset / styled / aligned in a single call.

### Consolidated Region Cons

* Changing header dimensions changes body cell / grid indexing.
    * Workaround: provide separate accessors for header / body, while still storing / providing an accessor for the combined table.

```python
import numpy
import toyplot.data
```

```python
canvas = toyplot.Canvas(width=400, height=400)
table = canvas.table(rows=10, columns=4, hrows=3)
table.grid.hlines[...] = "single"
table.grid.vlines[...] = "single"
table.header.grid.hlines[-1] = "double"
table.header.row(0).merge().data = "A"
table.header.cell(1,0,colspan=2).merge().data = "B"
table.header.cell(1,2,colspan=2).merge().data = "C"
table.header.cell(2,0).data = "1"
table.body.column(0).data = numpy.arange(0, 100, 10)
table.body.column(1).data = numpy.random.random(10)
table.column(0).align = "right"
table.column(1).format = toyplot.format.FloatFormatter("{:.2g}")
table.column(1).align = "separator"
```
