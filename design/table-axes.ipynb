```python
import numpy
import toyplot.data
```

```python
numpy.random.seed(1234)

data = toyplot.data.Table()
data["foo"] = numpy.arange(10)
data["bar"] = numpy.random.uniform(-5, 5, size=10)
data["baz"] = numpy.random.choice(["red", "green", "blue"], 10)
data["blah"] = numpy.repeat("", 10)
```

```python
numpy.random.seed(1234)
canvas = toyplot.Canvas(width=400, height=400)
canvas.text(200, 10, "Table Test")
table = canvas.table(data, colwidth={"blah":100})
table.cell_axes(1, 3).plot(numpy.sin(numpy.linspace(0, 10)))
table.cell_axes(2, 3).bars(numpy.random.uniform(0.1, 1, size=10))
table.cell_axes(3, 3).bars(numpy.random.choice([-1, 1], size=10), fill=numpy.random.choice(["red", "blue"], size=10))
table.cell_axes(4, 3).fill(1.5 + numpy.cos(numpy.linspace(0, 5)))
```
