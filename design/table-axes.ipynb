```python
import numpy
import toyplot.data
```

```python
numpy.random.seed(1234)

data = toyplot.data.Table()
data["foo"] = numpy.arange(0, 144, 15)
data["bar"] = numpy.random.normal(scale=100, size=10)
data["baz"] = numpy.random.choice(["red", "green", "blue"], 10)
data["blah"] = numpy.repeat("", 10)
```

```python
numpy.random.seed(1234)
canvas = toyplot.Canvas(width=400, height=400)
table = canvas.table(data)
table.column("foo").offset = -25
table.column("bar").offset = -5
table.column("baz").offset = 17
table.column(0).style = {"fill":"red"}
table.column("blah").header.content = "Blah"
table.column("blah").header.style = {"fill":"orange"}
table.row(5).style = {"font-weight":"bold", "fill":"blue"}
table.cell(5, 1).style = {"fill":"green"}
table.cell(8, 2).content = u"\u2937 t"
table.hlines[1,...] = "double"
table.vlines[...,1] = "single"
table.cell(0, 3).axes().plot(numpy.sin(numpy.linspace(0, 10)))
table.cell(1, 3).axes().bars(numpy.random.uniform(0.1, 1, size=10))
table.cell(2, 3).axes().bars(numpy.random.choice([-1, 1], size=30), fill=numpy.random.choice(["red", "blue"], size=30))
table.cell(3, 3, rowspan=2).axes().fill(3 + numpy.cos(numpy.linspace(0, 5)) + numpy.sin(numpy.linspace(0, 20)))
```

```python
import toyplot.pdf
toyplot.pdf.render(canvas, "test.pdf")
```
