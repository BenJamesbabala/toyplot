```python
import numpy
import toyplot.data
```

```python
data = toyplot.data.Table()
data["a"] = numpy.arange(10)
data["b"] = data["a"] ** 2
data["c"] = numpy.repeat("", 10)
```

```python
numpy.random.seed(1234)
canvas = toyplot.Canvas(width=300, height=600)
table = canvas.table(data)
table.cell_axes(1, 2).plot(numpy.sin(numpy.linspace(0, 10)))
table.cell_axes(2, 2).bars(numpy.random.uniform(0.1, 1, size=10))
table.cell_axes(3, 2).bars(numpy.random.choice([-1, 1], size=10), fill=numpy.random.choice(["red", "blue"], size=10))
table.cell_axes(4, 2).fill(1.5 + numpy.cos(numpy.linspace(0, 5)))
```
