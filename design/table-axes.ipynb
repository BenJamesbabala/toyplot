```python
import numpy
import toyplot.data
```

```python
canvas = toyplot.Canvas(width=400, height=400)
table = canvas.table(10, 4, title="Quarterly Report")
table.grid.hlines[...] = "single"
table.grid.vlines[...] = "single"
table.column(0).data = numpy.arange(0, 100, 10)
table.column(0).align = "right"
table.column(1).data = numpy.random.random(10)
table.column(1).format = toyplot.format.FloatFormatter("{:.2g}")
table.column(1).align = "separator"
table.header().cell(0,0).data = "A"
table.header().cell(0,1).data = "B"
```

```python
numpy.random.seed(1234)

data = toyplot.data.Table()
data["foo"] = numpy.arange(0, 144, 15)
data["bar"] = numpy.random.normal(scale=100, size=10)
data["baz"] = numpy.random.choice(["red", "green", "blue"], 10)
data["blah"] = numpy.repeat("", 10)
data
```

```python
canvas = toyplot.Canvas(width=400, height=400)
table = canvas.table(data, title="Quarterly Report")
table.grid.hlines[...] = "single"
table.grid.vlines[...] = "single"
table.cell(0, 3, rowspan=6).merge().data = "Foo"
table.cell(6, 3, rowspan=4).merge().data = "Bar"
```

```python
numpy.random.seed(1234)
canvas = toyplot.Canvas(width=400, height=420)
table = canvas.table(data)
table.column(0).offset = -25
table.column(1).offset = -5
table.column(2).offset = 17
table.column(0).style = {"fill":"red"}
#table.column(3).header.content = "Blah"
#table.column(3).header.style = {"fill":"orange"}
table.row(5).style = {"font-weight":"bold", "fill":"blue"}
table.cell(5, 1).style = {"fill":"green"}
table.cell(8, 2).data = u"\u2937 t"
table.grid.hlines[1,...] = "double"
table.grid.vlines[...,1] = "single"
table.grid.style = {"stroke":"gray"}
table.grid.separation = 3
table.cell(0, 3).axes().plot(numpy.sin(numpy.linspace(0, 10)))
table.cell(1, 3).axes().bars(numpy.random.uniform(0.1, 1, size=10))
table.cell(2, 3).axes().bars(numpy.random.choice([-1, 1], size=30), fill=numpy.random.choice(["red", "blue"], size=30))
table.cell(3, 3, rowspan=2).axes().fill(3 + numpy.cos(numpy.linspace(0, 5)) + numpy.sin(numpy.linspace(0, 20)))
```

```python
import toyplot.pdf
toyplot.pdf.render(canvas, "test.pdf")
```
