```python
import numpy
import toyplot.data
```

```python
numpy.random.seed(1234)

data = toyplot.data.Table()
data["foo"] = numpy.arange(10)
data["bar"] = numpy.random.normal(scale=100, size=10)
data["baz"] = numpy.random.choice(["red", "green", "blue"], 10)
data["blah"] = numpy.repeat("", 10)
```

```python
numpy.random.seed(1234)
canvas = toyplot.Canvas(width=400, height=400)
table = canvas.table(data)
table.column("baz").justify = "left"
table.column("baz").offset = 15
#table.column("bar").offset = 0
#table.column("bar").format = toyplot.format.FloatFormatter(format="%.4g")
table.column("blah").width = 120
table.cell(1, 3).axes().plot(numpy.sin(numpy.linspace(0, 10)))
table.cell(2, 3).axes().bars(numpy.random.uniform(0.1, 1, size=10))
table.cell(3, 3).axes().bars(numpy.random.choice([-1, 1], size=30), fill=numpy.random.choice(["red", "blue"], size=30))
table.cell(4, 3, rowspan=3).axes().fill(3 + numpy.cos(numpy.linspace(0, 5)) + numpy.sin(numpy.linspace(0, 20)))
```
