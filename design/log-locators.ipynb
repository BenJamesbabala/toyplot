```python
import numpy
x = numpy.linspace(.001, 1000000000)
```

```python
import toyplot.locator
locator = toyplot.locator.PositiveLog(format="{:.1e}")
```

```python
canvas = toyplot.Canvas()
axes = canvas.axes(xscale="log", yscale="log")
axes.x.ticks.labels.style = {"font-size":"18px"}
axes.y.ticks.labels.style = {"font-size":"20px"}
axes.plot(x, x, marker="o")
```

```python
import toyplot.pdf
toyplot.pdf.render(canvas, "test.pdf")
```
