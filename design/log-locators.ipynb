```python
import numpy
x = numpy.linspace(-1000, 0.01)
```

```python
import toyplot.locator
locator = toyplot.locator.Log()
```

```python
canvas = toyplot.Canvas()
axes = canvas.axes(xscale="log", yscale="log", xticklocator=locator)
axes.x.ticks.labels.style = {"font-size":"15px"}
axes.y.ticks.labels.style = {"font-size":"15px"}
axes.plot(x, x, marker="o")
```

```python
import toyplot.pdf
toyplot.pdf.render(canvas, "test.pdf")
```
