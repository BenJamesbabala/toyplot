```python
import numpy
x = numpy.linspace(-100, 100)
x
```

```python
import toyplot
canvas = toyplot.Canvas()
axes = canvas.axes(xscale=("log", 3), yscale=("log", 3))
axes.x.ticks.labels.style = {"font-size":"15px"}
axes.y.ticks.labels.style = {"font-size":"15px"}
axes.plot(x, x, marker="o")
```

```python
import toyplot.pdf
toyplot.pdf.render(canvas, "test.pdf")
```
