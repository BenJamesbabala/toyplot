Axes text use-cases:

```python

# Plot one label
axes.text(1.4, 3.2, "Label")

# Plot multiple labels
axes.text(numpy.linspace(0, 1), numpy.sin(linspace(0, 1)), numpy.arange(50))

# Plot table
axes.text(table, x="x", y=["y1", "y2"], text=["t1", "t2"])

```

````python
import numpy
import toyplot.data
````

````python
canvas = toyplot.Canvas(width=800, height=400)
canvas.axes(grid=(1, 2, 0)).text(1.4, 3.2, "Label")
canvas.axes(grid=(1, 2, 1)).text(numpy.linspace(0, 2 * numpy.pi), numpy.sin(numpy.linspace(0, 2 * numpy.pi)), numpy.arange(50))
````

````python
table = toyplot.data.Table()
table["x"] = numpy.linspace(0, 2 * numpy.pi)
table["sin(x)"] = numpy.sin(table["x"])
table["cos(x)"] = numpy.cos(table["x"])
table["t1"] = ["{:.3g}".format(value) for value in table["sin(x)"]]
table["t2"] = ["{:.3g}".format(value) for value in table["cos(x)"]]
````

````python
table.rows(numpy.index_exp[0:10])
````

````python
canvas = toyplot.Canvas()
axes = canvas.axes()
#axes.text(table["x"], table["sin(x)"], table["t1"])
axes.text(table, x="x", y=["sin(x)", "sin(y)"], text=["t1", "t2"])
````
