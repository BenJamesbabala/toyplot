Axes text use-cases:

```python

# Plot one label
axes.text(1.4, 3.2, "Label")

# Plot multiple labels
axes.text(numpy.linspace(0, 1), numpy.sin(linspace(0, 1)), numpy.arange(50))

# Plot table
axes.text(table, x="x", y=["y1", "y2"], text=["t1", "t2"])

```

````python
import numpy
import toyplot.data

x = numpy.linspace(0, 2 * numpy.pi)
y = numpy.sin(x)
````

````python
canvas = toyplot.Canvas(width=800, height=400)
axes = canvas.axes(grid=(1, 2, 0))
axes.text(1.4, 3.2, "Label 1")
axes.text(1.6, 3.0, "Label 2")
canvas.axes(grid=(1, 2, 1)).text(x, y, numpy.arange(50), annotation=False)
````
