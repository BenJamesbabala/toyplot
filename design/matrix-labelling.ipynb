```python
import numpy
import toyplot.color
```

```python
numpy.random.seed(1234)
data = numpy.random.random((16, 16))

canvas = toyplot.Canvas(width=800, height=400)

canvas.matrix(
    data,
    label="Matrix Visualization",
    tlabel="Stuff labelled on top",
    blabel="Stuff labelled on the bottom",
    llabel="Left-side labelling",
    rlabel="Bringing-up the right-hand side",
    step=1,
    grid=(1, 2, 0),
    gutter=30);

canvas.matrix(
    (data, toyplot.color.diverging("BlueRed", data.min(), data.max())),
    label="Matrix Visualization",
    step=1,
    rlocator=toyplot.locator.Explicit([0, 5, 7], ["First", "Fifth", "Seventh"]),
    blocator=toyplot.locator.Explicit([1, 3, 7, 9, 12]),
    grid=(1, 2, 1),
    gutter=30);
```

```python
import numpy
import toyplot.color
data = numpy.random.normal(size=(16, 16))
palette = toyplot.color.brewer("BlueRed")
canvas = toyplot.Canvas(width=600)
matrix = canvas.matrix((data, palette), label="My Matrix")
axes = canvas.axes(show=False)
axes.coordinates.show = False
axes.colorbar(palette=palette, min=data.min(), max=data.max(), label="Color Scale")
```
