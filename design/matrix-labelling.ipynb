```python
import numpy
import toyplot
```

```python
numpy.random.seed(1234)
data = numpy.random.random((16, 16))

canvas = toyplot.Canvas(width=800, height=400)

canvas.matrix(
    data,
    label="label",
    tlabel="Stuff labelled on top",
    blabel="Stuff labelled on the bottom",
    llabel="Left-side labelling",
    rlabel="Bringing-up the right-hand side",
    step=1,
    grid=(1, 2, 0),
    gutter=30);

canvas.matrix(
    data,
    label="label",
    step=1,
    rshow=True,
    rlocator=toyplot.locator.Explicit([0, 5, 7], ["First", "Fifth", "Seventh"]),
    grid=(1, 2, 1),
    gutter=30);
```

```python
canvas, axes, mark = toyplot.plot(numpy.linspace(0, 1) ** 2, width=300)
axes.x.show = True
axes.x.ticks.locator = toyplot.locator.Null()
axes.x.label.text = "Foo"
```
