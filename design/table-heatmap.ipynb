```python
import numpy
import toyplot
```

```python
colormap = toyplot.color.diverging("BlueRed", 0, 1)

data = numpy.array(["Tim", "Dick", "Harry", "June", "Amy", "Bev"])
columns = 52

canvas = toyplot.Canvas(width=1000, height=300)
table = canvas.table(rows=len(data), columns=columns + 1, title="Vacation Use")
table.column(0).data = data
table.column(0).width = 75
#table.grid.vlines[...] = "single"
table.gaps.rows[1:-1] = 10
#table.gaps.columns[0::7] = 5
for i in range(len(data)):
    for j in range(columns):
        color = colormap.css(numpy.random.normal(0.5, 0.1))
        table.cell(i, j + 1).bstyle = {"stroke":"none","fill":color}
```
